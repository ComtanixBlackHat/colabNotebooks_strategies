```python
import vectorbt as vbt
import yfinance as yf

data = yf.download("AAPL", start="2020-01-01", end="2024-01-01")["Close"]

rsi = vbt.RSI

signals = vbt.If(rsi < 30, 1, vbt.If(rsi > 70, -1, 0))


portfolio = vbt.Portfolio.from_signals(
    data=data,
    signals=signals,
    cash=10000,
    cost=0.001,
)


print(portfolio.stats())

portfolio.plot(style="line").add_span(signals == 1, color="green", alpha=0.3, label="Buy").add_span(signals == -1, color="red", alpha=0.3, label="Sell")
```
