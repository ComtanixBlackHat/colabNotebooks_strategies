```python
import datetime
datetime.datetime.

def convert_to_minutes(time_str):

    time_mapping = {
        'm': 1,
        'h': 60,
        'd': 1440,
        'w': 10080,
        'M': 43200
    }


    num = int(time_str[:-1])
    unit = time_str[-1]


    return num * time_mapping[unit]

print(type(convert_to_minutes("1m")))  # Output: 1
print(convert_to_minutes("3m"))  # Output: 3
print(convert_to_minutes("5h"))  # Output: 60
print(convert_to_minutes("3d"))  # Output: 1440
print(convert_to_minutes("1w"))  # Output: 10080
print(convert_to_minutes("1M"))  # Output: 43200
```
