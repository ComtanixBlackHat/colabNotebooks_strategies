```python
import pandas as pd


df = pd.read_csv('btceth_result.csv')
timestamps = df['Timestamp']


time_diffs_ms = timestamps.diff().dropna()


time_diffs_s = time_diffs_ms / 1000

average_time_diff_s = time_diffs_s.mean()
print(f"average time difference = {average_time_diff_s} secounds")
first_timestamp = df['Timestamp'].iloc[0]
second_timestamp = df['Timestamp'].iloc[-1]


totalTime = second_timestamp - first_timestamp


totalTime_s = (totalTime / 1000)/60
print(f"Total time = {totalTime_s} minutes")


df_eq2 = df[df['Equation'] == 'eq2 ']

df_eq1 = df[df['Equation'] == 'eq1 ']


total_eq2_count = len(df_eq2)
total_eq1_count = len(df_eq1)

# Print the total counts
print(f"Total USDT --> ETH --> BTC --> USDT: {total_eq2_count} (eq2)")
print(f"Total USDT --> BTC --> ETH --> USDT: {total_eq1_count} (eq1)")


time_diffs_ms_eq1 = df_eq1['Timestamp'].diff().dropna()


average_time_diff_ms_eq1 = time_diffs_ms_eq1.mean()


average_time_diff_s_eq1 = average_time_diff_ms_eq1 / 1000


print(f"Average time difference for eq1 USDT --> BTC --> ETH --> USDT: {average_time_diff_s_eq1} seconds")

# /////////////////////

time_diffs_ms_eq2 = df_eq2['Timestamp'].diff().dropna()


average_time_diff_ms_eq2 = time_diffs_ms_eq2.mean()


average_time_diff_s_eq2 = average_time_diff_ms_eq2 / 1000


print(f"Average time difference for eq2 USDT --> ETH --> BTC --> USDT: {average_time_diff_s_eq2} seconds")

lowest_conversion_rate = df['Conversion Rate'].min()
highest_conversion_rate = df['Conversion Rate'].max()


print(f"Lowest conversion rate: {lowest_conversion_rate}")
print(f"Highest conversion rate: {highest_conversion_rate}")
df.head()
```
