```python
import yfinance as yf
import vectorbt as vbt

# Define ticker symbol and date range
ticker = 'AAPL'
start_date = '2020-01-01'
end_date = '2024-01-01'

# Download historical price data
data = yf.download(ticker, start=start_date, end=end_date)

# Calculate RSI and Bollinger Bands (using default window sizes)
rsi = vbt.RSI(data.close)
bbands = rsi.bollinger_bands()

# Define clear and concise buy and sell signals
buy_signal = vbt.Cond(data.Close > 30, True, False) & vbt.Cond(data.Close < bbands.lower, True, False)
sell_signal = vbt.Cond(rsi >= 70, True, False) | vbt.Cond(data.Close > bbands.upper, True, False)

# Combine buy and sell signals into a single object for entry and exit points
entries = vbt.Portfolio.from_signals(buy_signal, sell_signal).oc()  # Open/Close from signals

# Run the strategy using closing price data
portfolio = entries.run(data.Close)

# Visualize the backtesting results
vbt.plotting.plot(portfolio, title="Bollinger Bands + RSI Backtest")
```
